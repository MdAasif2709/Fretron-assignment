```python
import matplotlib.pyplot as plt

flight_1 = [(1, 1), (2, 2), (3, 3)]
flight_2 = [(1, 1), (2, 4), (3, 2)]
flight_3 = [(1, 1), (4, 2), (3, 4)]

x1, y1 = zip(*flight_1)
x2, y2 = zip(*flight_2)
x3, y3 = zip(*flight_3)

plt.figure(figsize=(6, 6))
plt.plot(x1, y1, marker='o', color='yellow', label='Flight 1')
plt.plot(x2, y2, marker='o', color='blue', label='Flight 2')
plt.plot(x3, y3, marker='o', color='red', label='Flight 3')


plt.grid(True)
plt.legend()
plt.title('Flight Paths')
plt.xlabel('X-coordinate')
plt.ylabel('Y-coordinate')
plt.show()
```

```python
def distribute_apples(apple_weights):
    apple_weights.sort(reverse=True)
   
    total_weight = sum(apple_weights)
    target_weights = {
        'Ram': total_weight * 0.5,
        'Sham': total_weight * 0.3,
        'Rahim': total_weight * 0.2
    }
    
    distribution = {'Ram': [], 'Sham': [], 'Rahim': []}
    current_weights = {'Ram': 0, 'Sham': 0, 'Rahim': 0}
    
    for apple in apple_weights:
        added = False
        for person in ['Ram', 'Sham', 'Rahim']:
            if current_weights[person] + apple <= target_weights[person]:
                distribution[person].append(apple)
                current_weights[person] += apple
                added = True
                break
        
        if not added:
            distribution['Rahim'].append(apple)
            current_weights['Rahim'] += apple
    
    return distribution
def main():
    print("Enter apple weights in grams (-1 to stop):")
    apple_weights = []
    
    while True:
        weight = int(input())
        if weight == -1:
            break
        apple_weights.append(weight)
    
    distribution_result = distribute_apples(apple_weights)
    
    
    print("\nDistribution Result:")
    for person in distribution_result:
        print(f"{person} : {', '.join(map(str, distribution_result[person]))}")

if __name__ == "__main__":
    main()
```

```python
def simulate_castle_improved(soldiers, castle):
    paths = []
    directions = [(0, 1), (1, 0), (0, -1), (-1, 0)]  # right, down, left, up

    def move_castle(current_position, direction_index, visited, path):
        x, y = current_position
        dx, dy = directions[direction_index]

        while 1 <= x + dx <= 9 and 1 <= y + dy <= 9:
            x, y = x + dx, y + dy
            path.append((x, y))
            if (x, y) in soldiers:
                soldiers.remove((x, y))
                visited.add((x, y))

                # Turn left
                new_direction_index = (direction_index + 1) % 4
                move_castle((x, y), new_direction_index, visited, path)
                break

        if (x, y) == castle and path not in paths:
            paths.append(path.copy())

    for direction_index in range(4):
        move_castle(castle, direction_index, set(), [castle])

    return paths

soldiers = {(1, 1), (8, 9), (1, 9), (4, 1), (4, 2), (4, 8), (2, 6), (5, 6), (8, 2), (5, 9), (2, 8)}
castle = (1, 2)

unique_paths = simulate_castle_improved(soldiers, castle)
unique_paths
print(f"Number of unique paths: {len(unique_paths)}\n")

for i, path in enumerate(unique_paths, start=1):
    print(f"Path {i}:")
    for step in path:
        print(step, end=" -> ")
    print("Home\n")

```
